```python
import matplotlib.pyplot as plt
import requests
```

```python


# Define the URL
url = "https://stc.mapotic.com/api/v1/poi/358/move/?page_size=10000"

# Make a request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the response JSON data
    data = response.json()['results']

    # Print out the data or work with it as needed
    print(data)
else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")
```

```python
lat = []
lng = []
distance = []
duration = []
for i in data:
    lat.append(i['data']['Lat'])
    lng.append(i['data']['Lng'])
    distance.append(i['data']['Distance'])
    duration.append(i['data']['Duration'])
lng.reverse()
lat.reverse()
distance.reverse()
duration.reverse()
```

```python
len(lat)
```

```python
plt.plot(lng,lat)
plt.show()
```
